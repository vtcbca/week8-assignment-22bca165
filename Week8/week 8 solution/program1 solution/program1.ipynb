```python
df = pd.read_csv("product.csv")
```

```python
df['Total_Sell'] = df[['Jan', 'Feb', 'Mar', 'Apr', 'may', 'Jun']].sum(axis=1)
```

```python
df['Average_Sell'] = df[['Jan', 'Feb', 'Mar', 'Apr', 'may', 'Jun']].mean(axis=1)
```

```python
plt.figure(figsize=(10, 6))
```

```python
for product in df['prod_name']:
    plt.plot(df.columns[1:7], df[df['prod_name'] == product].iloc[:, 1:7].values.flatten(), marker='o', label=product)
    plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Product Sales Analysis for 6 Months')
plt.legend()
plt.grid(True)
```

```python
df.to_csv('sell_analysis.csv', index=False)
```
